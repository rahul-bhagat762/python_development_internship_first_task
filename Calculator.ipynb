```python
def add(a,b):
    """Return the sum of two numbers."""
    return a+b
def subtract(a,b):
    """Return the subtract of two numbers."""
    return a-b
def multiply(a,b):
    """Return the multiplication of two numbers."""
    return a*b
def divide(a,b):
    """Return the sum of two numbers."""
    if b == 0:
        return "Error: Division by zero is undefined."
    return a / b









def evaluate_expression(exp):
    """Evaluates a basic arithmetic expression and returns the result."""
    try:
        # if "/ 0" in expression or "/0" in expression:
        #     return "Error: Division by zero is undefined."

        result = eval(exp,{'__builtins__': None},{"+":add , "-": subtract ,"*":multiply , "/":divide})
        return result
    except(SyntaxError,NameError,TypeError):
        return "Error : Invalid input,please enter a valid Expression."
    except(ZeroDivisionError):
        return "Error: Division by zero is undefined."









def main():
    """main function for the command line calculator interface"""
    print('''Welcome to the python text-Based Calculator \n
    you can perform basic operations like addition(+),subtraction(-),multiplication(*) and division(/).''')
    print("Type 'exit' to quit the calculator.")

    while(True):
        user_input = input("Enter an expression like '(a(oprand) operator b(operand))':").strip()
        if user_input.lower() == 'exit':
            print("Thank you for using the calculator. Goodbye!")
            break
        result = evaluate_expression(user_input)
        print(f"Result: {result}")




if __name__ == '__main__':
    main()
```
